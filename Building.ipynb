- Mean Squared Error (MSE): It measures the average squared difference between the predicted and actual values. A lower MSE indicates better model performance, with a value of 0 indicating a perfect fit.

- Mean Absolute Error (MAE): It measures the average absolute difference between the predicted and actual values. Like MSE, a lower MAE suggests better model performance. MAE is less sensitive to outliers compared to MSE.

- R-squared Score: Also known as the coefficient of determination, it represents the proportion of the variance in the dependent variable that can be explained by the independent variables. R-squared ranges from 0 to 1, with a higher value indicating a better fit. A score of 1 means the model explains all the variance, while a score of 0 means the model provides no improvement over a constant mean.
